```python
data = {
    'a':np.arange(50),
    'b':np.arange(50),
    'd':np.random.randn(50)
}
data['b'] = data['a']+10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
data
```

```python
plt.scatter('a','b',data=data)
```

```python
plt.scatter('a','b',data=data,c='c',s='d',marker='^')
plt.minorticks_on()
```

```python
names= ['group_a','group_b','group_c']
values=[1,10,100]
plt.figure(figsize=(9,3))
plt.subplot(131)
plt.bar(names,values) #막대
plt.subplot(132)
plt.scatter(names,values) # 점
plt.subplot(133)
plt.plot(names,values) #선 
```

```python
#figure이 그냥..음 .. id같은 느낌이네
#첫번쨰 , 두번째 , 세번쨰는 첫번쨰+두번째
#첫번째  
plt.figure(1)
plt.subplot(211)  #2행의 1열의 1열
plt.plot([1,2,3])
plt.subplot(212) # 2행의 1열의 2열
plt.plot([1,2,3])
#두번째
plt.figure(2)  #
plt.plot([1,2,3])
#세번째
plt.figure(1)
plt.subplot(211)
```
