```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784')
```

```python
mnist
```

```python
mnist.data.shape
```

```python
mnist.target.shape
```

```python
#pickle 은 파이썬안에서 저장할때 좋음. 이유는 그대로 저장했다가 그대로 불러올수 있음
```

```python
import pickle 
```

```python
with open('mnist.txt','wb') as f: 
    pickle.dump(mnist,f)
```

```python
with open('mnist.txt','rb') as f:
    mn = pickle.load(f)
```

```python
mn
```

```python
mn.target_names
```

```python
mn.target[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(mn.data,mn.target,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X_train,Y_train) #학습
```

```python
pred = clf.predict(X_test) #예측
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,pred) #정답, 예측값    을 넣어준다.
```
